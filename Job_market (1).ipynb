**Explain Columns**

year
Represents the year in which the job market data was recorded. It is used to analyze yearly changes in salaries, job demand, and overall hiring trends over time.

month
Indicates the numeric month within the year. It helps identify monthly patterns and short-term fluctuations in hiring activity.

month_name
Shows the name of the month, mainly used to improve readability and clarity in visualizations related to monthly trends.

quarter
Represents the quarter of the year (Q1–Q4). It is useful for comparing job market performance across different quarters.

season
Indicates the season during which the job data was recorded. It helps analyze seasonal effects on hiring and job availability.

job_title
Represents the job position or role. It is a core dimension used to analyze salaries, demand levels, and job availability across different roles.

industry
Shows the industry or sector associated with each job. It is used to compare hiring demand and salary levels across different industries.

experience_level
Indicates the level of experience required for the job, such as junior, mid-level, or senior. It helps assess how experience impacts salary and demand.

education_requirement
Represents the minimum educational qualification required for the job. It is used to study the relationship between education level and job opportunities.

salary_max_usd
Represents the maximum expected salary for the job in USD. It is used to evaluate compensation levels and compare salary ranges across jobs and regions.

salary_variability
Measures how much salaries vary within the same job role or industry. Higher variability indicates less salary stability or wider negotiation ranges.

salary_bracket
Categorizes salaries into ranges such as low, medium, and high. It simplifies salary distribution analysis and comparisons.

number_of_openings
Indicates the total number of available job openings. It is a direct measure of market demand for specific roles.

hiring_trend
Represents the hiring direction over time, such as increasing, stable, or decreasing. It helps identify growth or decline in employment demand.

is_high_demand
Indicates whether a job is classified as high demand. It helps quickly identify the most sought-after roles in the job market.

openings_level
Classifies the number of job openings into levels such as low, medium, or high. This makes demand comparisons easier and more intuitive.

region
Represents the broader geographical region of the job. It is used to analyze regional differences in salaries and hiring demand.

metro_city
Indicates whether the job is located in a major metropolitan area. It helps assess the impact of urban locations on job opportunities and salaries.

cost_region
Represents the cost-of-living level of the region. It is used to compare salaries against living expenses to evaluate real purchasing power.

remote_availability
Indicates whether the job supports remote or hybrid work. It helps analyze trends in flexible and remote employment.

job_market_pressure
Measures the level of pressure in the job market based on supply and demand. Higher pressure suggests fewer jobs relative to applicants.

competitiveness_score
Represents an overall score indicating how competitive a job role is. It reflects the intensity of competition among candidates for available positions.

**Cheak Outlier**

```python
from numpy._core import numeric
numeric = df.select_dtypes(include=np.number)
numeric_columns = numeric.columns
for col in df.columns:
  if col in numeric_columns:
    plt.figure(figsize=(10,5))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

**Visualization**

```python
salary_by_month = df.groupby('month')['salary_max_usd'].mean()

plt.figure(figsize=(10,5))
plt.plot(salary_by_month.index, salary_by_month.values)
plt.xlabel('month')
plt.ylabel('Average Max Salary (USD)')
plt.title('Average opening Trend Over month')
plt.show()
```

```python
salary_by_year = df.groupby('month')['number_of_openings'].mean()

plt.figure(figsize=(10,5))
plt.plot(salary_by_year.index, salary_by_year.values)
plt.xlabel('month')
plt.ylabel('Average opening (USD)')
plt.title('Average opening Trend Over month')
plt.show()
```

```python
salary_by_year = df.groupby('job_title')['job_market_pressure'].mean()

plt.figure(figsize=(30,10))
plt.bar(salary_by_year.index, salary_by_year.values)
plt.xlabel('job_title')
plt.ylabel('job_pressure')
plt.title('job_pressure for job')
plt.show()
```

**Transformation**

transfor object data to numeric

```python
from sklearn.preprocessing import LabelEncoder

category_cols = df.select_dtypes(include='object').columns

labelencoder = LabelEncoder()

for col in category_cols:
    df[col] = labelencoder.fit_transform(df[col])

```
